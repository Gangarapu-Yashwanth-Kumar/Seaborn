```python
vis1 = sns.distplot(df["InternetUsers"]) # univariate analysis
plt.show(vis1) # it is advanced graph
```

```python
vis1 = sns.displot(df["InternetUsers"]) # t is removed. so distribution line is removed.
plt.show()
```

```python
vis2 = sns.boxplot(data = df, x= 'IncomeGroup', y = 'BirthRate')
plt.show()
```

```python
vis3 = sns.lmplot(data=df, x = "InternetUsers", y = "BirthRate")
plt.show() # the blue dots are each data present in the excel
```

```python

vis3 = sns.lmplot(data=df, x = "InternetUsers", y = "BirthRate", fit_reg = False)
plt.show() 
```
